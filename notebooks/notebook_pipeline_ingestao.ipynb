# Pipeline de Ingestão - FULL e INCREMENTAL

Este notebook executa o pipeline completo de ingestão de dados com suporte a:
- **Modo FULL**: Recarrega todos os dados
- **Modo INCREMENTAL**: Coleta apenas dados novos
- **Modo AUTO**: Decide automaticamente

## Como Usar

Execute a célula abaixo escolhendo o modo:
- `mode='full'` - Recarrega tudo
- `mode='incremental'` - Apenas novos dados
- `mode='auto'` - Decide automaticamente (padrão)

```python
# Configurar modo de execução
MODE = 'auto'  # Opções: 'full', 'incremental', 'auto'

print(f"Modo selecionado: {MODE.upper()}")
```
